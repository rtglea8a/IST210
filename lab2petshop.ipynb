Lab 2: Pet Shop ER Diagram

  The Entity Relationship Diagram (ER), as shown above, has been constructed with the purpose of creating a helpful database for a local pet shop business.  Following specific aspects of this business, I have constructed six different tables titled: Enclosure, Food, Customer, Loyalty, Transaction (Pets), and Transaction (Food).  With these tables, I can show the relationships between them and how they function together in a database model.  Starting with the Customer entity, this table maintains four connections, with each being different. Between the Customer and the Transaction (Pets) entities, a one to many relationship exists, with one customer purchasing multiple pets.  Between the Customer and the Transaction (Food) entities, another one to many relationship exists, with one customer buying multiple sources of food.  In addition to a customer making purchases, there is also a loyalty program that allows that customer to earn reward points.  There is a one to many relationship present here as one customer earns multiple reward points. To the left side of the ER diagram is the entity titled Enclosure.  When a customer is looking for a specific animal, they visit one of three different enclosures.  Between the Customer and Enclosure entities, a one to one relationship exists, with one customer visiting one enclosure.  The last entity in this pet shop business model is the food entity. This table is connected to the enclosure table in a many to many relationship as multiple amounts of food are stored in multiple enclosures.  This business model that I have created for the local pet shop effectively demonstrates and portrays the active relationships between all six entities. With this information, we can better understand how a business works and can also use this information to improve its overall functionality.
